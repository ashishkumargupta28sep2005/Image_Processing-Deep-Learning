```python
import cv2 as cv
cam = cv.VideoCapture(0)

while True:

    _, img = cam.read()
    
    img = cv.flip(img, 1) 
    rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
    
    #RGB Extraction
    r = img[:,:,2]
    g = img[:,:,1]
    b = img[:,:,0]
    
    #HSV Extraction
    h = hsv[:,:,0]
    s = hsv[:,:,1]
    v = hsv[:,:,2]

    #RGB Frame Display
    cv.imshow("Frame",img)
    cv.imshow("Red"  ,r)
    cv.imshow("Green",g)
    cv.imshow("Blue" ,b)
  
    #HSV Frame Display    
    cv.imshow("HSV"  ,hsv)
    cv.imshow("Hue"  ,h)
    cv.imshow("Saturation",s)
    cv.imshow("Value",v)
    
    key = cv.waitKey(20)
    
    
    if (key == 27):
        cam.release()
        break

cam.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

cam = cv.VideoCapture(0)  # Open webcam

while True:
    _, img = cam.read()  # Capture frame
    img = cv.flip(img, 1)  # Flip image horizontally
    
    hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)  # Convert to HSV

    # Split RGB channels (Corrected: OpenCV uses BGR format)
    b, g, r = cv.split(img)  # Extract B, G, R channels
    
    # Create colorized Red, Green, and Blue images
    red   = np.zeros_like(img)
    red[:, :, 2] = r  # Only Red channel active

    green = np.zeros_like(img)
    green[:, :, 1] = g  # Only Green channel active

    blue  = np.zeros_like(img)
    blue[:, :, 0] = b  # Only Blue channel active

    # Split HSV channels
    h, s, v = cv.split(hsv)

    # Apply colormap for better HSV visualization
    hue_colored = cv.applyColorMap(h, cv.COLORMAP_JET)
    sat_colored = cv.applyColorMap(s, cv.COLORMAP_JET)
    val_colored = cv.applyColorMap(v, cv.COLORMAP_JET)

    # Display frames
    cv.imshow("Frame", img)        # Original Frame
    cv.imshow("Red Channel", red)  # Red-colored Image
    cv.imshow("Green Channel", green)  # Green-colored Image
    cv.imshow("Blue Channel", blue)  # Blue-colored Image

    # Display HSV channels
    cv.imshow("HSV", hsv)
    cv.imshow("Hue", hue_colored)
    cv.imshow("Saturation", sat_colored)
    cv.imshow("Value", val_colored)

    key = cv.waitKey(20) & 0xFF  # Capture key press

    if key == 27:  # If 'Esc' key is pressed
        cam.release()
        cv.destroyAllWindows()  # Close all windows
        break
```
